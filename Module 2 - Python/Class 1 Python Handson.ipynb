### **if else statements**

```python
salary = int(input('enter the salary you earn'))  #casting
if salary >= 70000:
    print("great salary! high earner")
elif salary >= 50000:
    print('average salary!good job')
else:
    print('salary is bad')
```

```python
def tax_brakcet(salary):
    if salary >= 70000:
        return "30%"
    elif salary >=5000:
        return "20%"
    else:
        return "10"
```

```python
salary = int(input('enter the salary you earn')) 
print(tax_brakcet(salary))
```
